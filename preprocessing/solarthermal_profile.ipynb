```python
import pandas as pd
import openpyxl

# Replace 'your_file.xlsx' with the path to your Excel file
PATH_IN = '../data/misc/profiles/'
PATH_OUT = '../data/input/profiles/'

PROFILE = '250116_Lastänge_Alt-Lössnig.xlsx'

YEARS = [2028, 2029, 2030, 2035, 2040, 2045, 2050]

year_col_map = {
    2028: ("Datum", "Zeit", "Erzeugung Solarthermie_MW"),
    2029: ("Datum.1", "Zeit.1", "Erzeugung Solarthermie_MW.1"),
    2030: ("Datum.2", "Zeit.2", "Erzeugung Solarthermie_MW.2"),
    2035: ("Datum.3", "Zeit.3", "Erzeugung Solarthermie_MW.3"),
    2040: ("Datum.4", "Zeit.4", "Erzeugung Solarthermie_MW.4"),
    2045: ("Datum.5", "Zeit.5", "Erzeugung Solarthermie_MW.5"),
    2050: ("Datum.6", "Zeit.6", "Erzeugung Solarthermie_MW.6"),
}

# Read the Excel file
df_profile_min = pd.read_excel(PATH_IN + PROFILE, sheet_name='ST Süd_min')
df_profile_max = pd.read_excel(PATH_IN + PROFILE, sheet_name='ST Süd_max')


# Extract the data for each year for the minimum profile
for year, (date_col, time_col, gen_col) in year_col_map.items():
    df_year = df_profile_min[[date_col, time_col, gen_col]].copy()
    df_year.columns = ['date', 'time', 'value']
    df_year['value'] = df_year['value'].astype(str).str.replace(',', '.').astype(float)
    df_year = df_year.dropna(subset=['time'])
    df_year.index = range(1, len(df_year) + 1)
    df_year.index.name = 't'
    df_year = df_year.drop(columns=['date', 'time'])
    print(df_year.head())
    df_year.to_csv(PATH_OUT + f'min_solarthermal_profil_{year}.csv')


# Extract the data for each year for the maximum profile
for year, (date_col, time_col, gen_col) in year_col_map.items():
    df_year = df_profile_max[[date_col, time_col, gen_col]].copy()
    df_year.columns = ['date', 'time', 'value']
    df_year['value'] = df_year['value'].astype(str).str.replace(',', '.').astype(float)
    df_year = df_year.dropna(subset=['time'])
    df_year.index = range(1, len(df_year) + 1)
    df_year.index.name = 't'
    df_year = df_year.drop(columns=['date', 'time'])
    print(df_year.head())
    df_year.to_csv(PATH_OUT + f'max_solarthermal_profil_{year}.csv')



```

```python
import pandas as pd
import openpyxl
import os


def normalize_solar_thermal_profile(profile):
    # Normalize the profile
    
    column = "value"
    
    max_value = profile[column].max()

    profile[column] = profile[column] / max_value
    return profile, max_value


file = 'dummy_solarthermal_profil.csv'

input_file = '../data/input/profiles/dummy/'

output_file = '../data/input/profiles_normalized/'
os.makedirs(output_file, exist_ok=True)


profile = pd.read_csv(input_file + file, index_col='t')

norm_profile, max_value = normalize_solar_thermal_profile(profile)

norm_profile.to_csv(output_file+file, index=True)

print(f"Profil normalized. Max Value: {max_value}")
print(f"Normalized profile saved to {output_file+file}")


```

```python
import gurobi_logtools as glt
from mpl_interactions import ioff, panhandler, zoom_factory
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
results_3600_1 = glt.parse('../data/postprocessing/240212_analyze_gap/solver_3600_1.log')
results_3600_2 = glt.parse('../data/postprocessing/240212_analyze_gap/solver_3600_2.log')
results_3600_3 = glt.parse('../data/postprocessing/240212_analyze_gap/solver_3600_3.log')
results_3600_4 = glt.parse('../data/postprocessing/240212_analyze_gap/solver_3600_4.log')
results_3600_5 = glt.parse('../data/postprocessing/240212_analyze_gap/solver_3600_5.log')
results_rqe = glt.parse('../data/postprocessing/240212_analyze_gap/solver_reduced_quadratic_equations.log')
log_progress_3600_1 = results_3600_1.progress()
log_progress_3600_2 = results_3600_2.progress()
log_progress_3600_3 = results_3600_3.progress()
log_progress_3600_4 = results_3600_4.progress()
log_progress_3600_5 = results_3600_5.progress()
log_progress_rqe = results_rqe.progress()
print(log_progress_3600_5)
```

```python
with plt.ioff():
    fig, ax = plt.subplots(dpi=120)
    ax.plot(log_progress_3600_1['Time'], log_progress_3600_1['Gap'], label='tl_3600_1')
    ax.plot(log_progress_3600_2['Time'], log_progress_3600_2['Gap'], label='tl_3600_2')
    ax.plot(log_progress_3600_3['Time'], log_progress_3600_3['Gap'], label='tl_3600_3')
    ax.plot(log_progress_3600_4['Time'], log_progress_3600_4['Gap'], label='tl_3600_4')
    ax.plot(log_progress_3600_5['Time'], log_progress_3600_5['Gap'], label='tl_3600_5')
    ax.plot(log_progress_rqe['Time'], log_progress_rqe['Gap'], label='tl_rqe')
    ax.grid()
    ax.set_ylabel('gap [-]')
    ax.set_xlabel('optimization time [s]')
    ax.legend()
disconnect_zoom = zoom_factory(ax)
pan_handler = panhandler(fig)
display(fig.canvas)
```

```python
fig, ax = plt.subplots(dpi=120, figsize=(17, 7.5))

ax.plot(log_progress_3600_1['Time'], log_progress_3600_1['Gap'], label='tl_3600_1')
ax.plot(log_progress_3600_2['Time'], log_progress_3600_2['Gap'], label='tl_3600_2')
ax.plot(log_progress_3600_3['Time'], log_progress_3600_3['Gap'], label='tl_3600_3')
ax.plot(log_progress_3600_4['Time'], log_progress_3600_4['Gap'], label='tl_3600_4')
ax.plot(log_progress_3600_5['Time'], log_progress_3600_5['Gap'], label='tl_3600_5')
ax.plot(log_progress_rqe['Time'], log_progress_rqe['Gap'], label='tl_rqe')
ax.grid()
ax.set_ylabel('gap [-]')
ax.set_xlabel('optimization time [s]')
ax.legend()
```

```python
with plt.ioff():
    fig, axes = plt.subplots(ncols=1, nrows=3, dpi=120, figsize=(17, 7.5))
    axes[0].plot(log_progress_3600_1['Time'], log_progress_3600_1['Gap'], label='tl_3600_1')
    axes[0].plot(log_progress_3600_2['Time'], log_progress_3600_2['Gap'], label='tl_3600_2')
    axes[0].plot(log_progress_3600_3['Time'], log_progress_3600_3['Gap'], label='tl_3600_3')
    axes[0].plot(log_progress_3600_4['Time'], log_progress_3600_4['Gap'], label='tl_3600_4')
    axes[0].plot(log_progress_3600_5['Time'], log_progress_3600_5['Gap'], label='tl_3600_5')

    axes[1].plot(log_progress_3600_1['Time'], log_progress_3600_1['Incumbent'])
    axes[1].plot(log_progress_3600_2['Time'], log_progress_3600_2['Incumbent'])
    axes[1].plot(log_progress_3600_3['Time'], log_progress_3600_3['Incumbent'])
    axes[1].plot(log_progress_3600_4['Time'], log_progress_3600_4['Incumbent'])
    axes[1].plot(log_progress_3600_5['Time'], log_progress_3600_5['Incumbent'])

    axes[2].plot(log_progress_3600_1['Time'], log_progress_3600_1['BestBd'])
    axes[2].plot(log_progress_3600_2['Time'], log_progress_3600_2['BestBd'])
    axes[2].plot(log_progress_3600_3['Time'], log_progress_3600_3['BestBd'])
    axes[2].plot(log_progress_3600_4['Time'], log_progress_3600_4['BestBd'])
    axes[2].plot(log_progress_3600_5['Time'], log_progress_3600_5['BestBd'])

    axes[0].grid()
    axes[1].grid()
    axes[2].grid()

    axes[0].set_ylabel('gap [-]')
    axes[1].set_ylabel('incumbent [-]')
    axes[2].set_ylabel('best bound [-]')

    axes[2].set_xlabel('optimization time [s]')

    fig.legend(ncols=5, bbox_to_anchor=(0.99, 1.04))
    fig.tight_layout()
disconnect_zoom = zoom_factory(axes[0])
disconnect_zoom = zoom_factory(axes[1])
disconnect_zoom = zoom_factory(axes[2])
pan_handler = panhandler(fig)
display(fig.canvas)
```
